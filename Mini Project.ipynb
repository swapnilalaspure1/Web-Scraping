### Lets import the library

```python
# To convert our output into csv
import pandas as pd

# To scrape the data from website
from bs4 import BeautifulSoup

# To get a https request from websites
import requests
from requests import get
```

```python
url = 'https://www.flipkart.com/search?q=mobile+under+50000+rs&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
# Empty List
pro_name = []
pro_price = []
pro_rating = []
```

```python
for i in range(1,355):  # Adjust the range as per the actual number of pages available
    page_url = f'{url}&page={i}'
    r = requests.get(page_url)
    soup = BeautifulSoup(r.text, 'html.parser')
    master_container = soup.find_all('div', {'class': '_2kHMtA'})    
    
    if not master_container:
        break  # Break the loop if there is no data on the page

    for j in master_container:
        name = j.find('div', {'class': '_4rR01T'})
        price = j.find('div', {'class': '_30jeq3 _1_WHN1'})
        rating = j.find('div', {'class': '_3LWZlK'})

        if name and price:  # Check if name and price elements are found
            pro_name.append(name.text.strip())
            pro_price.append(price.text.strip())
        else:
            # If name or price is not found, we skip this entry
            continue

        if rating:
            pro_rating.append(rating.text.strip())
        else:
            pro_rating.append("NA")
```

```python
# Create the DataFrame
df = pd.DataFrame({"Product Name": pro_name, "Product Price": pro_price, "Product Rating": pro_rating})
df
```
