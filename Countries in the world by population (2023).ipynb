```python
# Importing Required Library
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
# Set the path to the ChromeDriver executable
chrome_driver_path = r"C:\Users\Swapnil Alaspure\Documents\MIT Exam 2nd\WebScraping\chromedriver.exe"

# Create ChromeOptions object
chrome_options = webdriver.ChromeOptions()

# Instantiate the Chrome WebDriver with the specified options
driver = webdriver.Chrome(options=chrome_options)
driver.maximize_window()
# Open the URL
driver.get("https://www.worldometers.info/world-population/population-by-country/")
```

```python
# Country

# Our list empty list
country = []

for i in range (1,235):
    name_xpath = '//*[@id="example2"]/tbody/tr[' + str(i) +']/td[2]'
    
    name = driver.find_element(By.XPATH, 
                               name_xpath).text
    country.append(name)

# Display the list
country
```

```python
# Population 2023

# Our list empty list
population = []

for i in range (1,235):
    population_xpath = '//*[@id="example2"]/tbody/tr[' + str(i) +']/td[3]'
    
    pop = driver.find_element(By.XPATH, 
                               population_xpath).text
    
    population.append(pop)

# Display the list
population
```

```python
# Yearly Change

# Our list empty list
YrsChange = []

for i in range (1,235):
    YrsChange_xpath = '//*[@id="example2"]/tbody/tr[' + str(i) +']/td[4]'
    
    change = driver.find_element(By.XPATH, 
                               YrsChange_xpath).text
    
    YrsChange.append(change)

# Display the list
YrsChange
```

```python
# Net Change

# Our list empty list
NetChange = []

for i in range (1,235):
    NetChange_xpath = '//*[@id="example2"]/tbody/tr[' + str(i) +']/td[5]'
    
    change = driver.find_element(By.XPATH, 
                               NetChange_xpath).text
    
    NetChange.append(change)

# Display the list
NetChange
```

```python
# Density (P/km^2)

# Our list empty list
Density = []

for i in range (1,235):
    Density_xpath = '//*[@id="example2"]/tbody/tr['+ str(i) +']/td[6]'
    
    den = driver.find_element(By.XPATH, 
                               Density_xpath).text
    
    Density.append(den)

# Display the list
Density
```

```python
# Land Area (km^2)

# Our list empty list
LandArea = []

for i in range (1,235):
    LandArea_xpath = '//*[@id="example2"]/tbody/tr['+ str(i) +']/td[7]'
    
    La = driver.find_element(By.XPATH, 
                               LandArea_xpath).text
    
    LandArea.append(La)

# Display the list
LandArea
```

```python
# Migrant (net)

# Our list empty list
Migrant = []

for i in range (1,235):
    Migrant_xpath = '//*[@id="example2"]/tbody/tr['+ str(i) +']/td[8]'
    
    migrant = driver.find_element(By.XPATH, 
                               Migrant_xpath).text
    
    Migrant.append(migrant)

# Display the list
Migrant
```

```python
# Fert.Rate

# Our list empty list
FertRate = []

for i in range (1,235):
    FertRate_xpath = '//*[@id="example2"]/tbody/tr['+ str(i) +']/td[9]'
    
    frtrate = driver.find_element(By.XPATH, 
                               FertRate_xpath).text
    
    FertRate.append(frtrate)

# Display the list
FertRate
```

```python
# Med. Age

# Our list empty list
MedAge = []

for i in range (1,235):
    MedAge_xpath = '//*[@id="example2"]/tbody/tr['+ str(i) +']/td[10]'
    
    medAge = driver.find_element(By.XPATH, 
                               MedAge_xpath).text
    
    MedAge.append(medAge)

# Display the list
MedAge
```

```python
# Urban Pop

# Our list empty list
UrbanPop = []

for i in range (1,235):
    UrbanPop_xpath = '//*[@id="example2"]/tbody/tr['+ str(i) +']/td[11]'
    
    urbanPop = driver.find_element(By.XPATH, 
                               UrbanPop_xpath).text
    
    UrbanPop.append(urbanPop.replace(' ',''))

# Display the list
UrbanPop
```

```python
# World Share

# Our list empty list
WorldShare = []

for i in range (1,235):
    WorldShare_xpath = '//*[@id="example2"]/tbody/tr['+ str(i) +']/td[12]'
    
    worldShare = driver.find_element(By.XPATH, 
                               WorldShare_xpath).text
    
    WorldShare.append(worldShare.replace(' ',''))

# Display the list
WorldShare
```

```python
# Creating Data Frame
Pop_Country = pd.DataFrame({'Country':country,
                             "Population":population,
                             "Yrs Change":YrsChange,
                             "Net Change":NetChange,
                             "Density (P/Km^2)":Density,
                             "Land Area (Km^2)":LandArea,
                             "Migrant":Migrant,
                            "Fert. Rate":FertRate,
                            "Med. Age":MedAge,
                            "Urban Population":UrbanPop,
                            "World Share %":WorldShare})
Pop_Country
```

```python
# To Close the driver
driver.close()
```

```python
# storing into the excel file 
Pop_Country.to_excel("Pop_Country.xlsx")
```
